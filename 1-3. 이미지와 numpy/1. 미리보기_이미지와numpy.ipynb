### [2] 이미지  제작
- 이미지는 가로 세로의 픽셀에 0~255의 값으로 구성되는 array(배열) 구조로 작성됨
- 흑백이미지는 한개의 픽셀에 0~255의 값으로 표현
    - [ 
       <br> [0,0,255,0], 
       <br> [0,200,150,20]
       <br> ]
- 컬러이미지는 한개의 픽셀에 3개의 채널이 혼합되어 표현됨
    - Red 0-255, Blue 0-255, Green 0-255
    -[
      <br> [  [0,0,0], [0,0,0], [255,255,255], [0,0,0] ],
      <br> [ [8,255,0], [0,200,0], [20,255,255], [0,0,0] ]
      <br> ]
