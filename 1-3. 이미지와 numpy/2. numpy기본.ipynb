###  행열(array)의 차원수 확인
- np.ndim 명령으로 진행함.
- r,g,b의 컬러이미지 한장은  기본 3차원 (행,열,rgb값)
- r,g,b,투명여부 의 png 컬러이미지 한장은 4차원(행,열,rgb값,투명여부)
- 그레이계열의 이미지한장은 2차원

```python
import numpy as np
a1=[1,2,3]
a2=[[1,2,3],[2,5,3]]
a3=[
    [[1,2,3],[2,5,3]],
    [[1,2,3],[2,5,3]]
]

a1,a2,a3=np.array(a1),np.array(a2),np.array(a3)
np.ndim(a1),np.ndim(a2),np.ndim(a3)
```

### 행열(array)의 차원을포함한 행,열수
- np.shape (1장의 이미지만 볼때는 전체사진갯수는출력되지 않음)
- 그레이계열은 (전체사진갯수,행갯수,열갯수)
- r,g,b 컬러계열은 (전체사진갯수,행갯수,열갯수,rgb갯수)
- 참고: ndim은 len(shape)임,  np.ndim(a)는 len(np.shape(a))
- 필수개념:  2개이상의 비정형데이터셋은 반드시 shape이 같아야함.
-    즉 사진2장을 갖고있는 텐서안의 각각이미지는 서로 shape이 같아야함.


```python
a1=[1,2,3]
a2=[[1,2,3],[2,5,3]]
a3=[
    [[1,2,3],[2,5,3]],
    [[1,2,3],[2,5,3]]
]

a1,a2,a3=np.array(a1),np.array(a2),np.array(a3)
np.shape(a1),np.shape(a2),np.shape(a3)
```

### 행열 구조 변환 -reshape
- 변경전이나 변경후나 전체 원소의 갯수는 동일해야함.
- np.reshape(변수,행,열,rgb값)
     - 참고, 리스트자료일때는 
        - a=[1,2,3,4,5,6] , np.reshape(a,(2,3))
    
- 넘파이변수.reshape(행,열,rgb값)
    - 참고, 넘파이 자료일때는 
        - a=[1,2,3,4,5,6] , a=np.array(a), a.reshape(2,3)

```python
a = np.arange(6)  #임의로 6개의 변수를 제작
print(type(a))
a, a.shape, a.reshape(6,1), a.reshape(1,6),a.reshape(2,3)
```

```python
a=[1,2,3,4,5,6]
np.reshape(a,(2,3))
```

## numpy연결
- vstack, hstack
- 여러개의 이미지를 한번에 볼때 많이 사용

```python
a=np.zeros((5,4,3))
b=np.zeros((5,4,3))
b[:,:,:]=255

plt.subplot(1,4,1)
plt.imshow(a)
plt.subplot(1,4,2)
plt.imshow(b)

plt.subplot(1,4,3)
plt.imshow(np.hstack([a,b]))

plt.subplot(1,4,4)
plt.imshow(np.vstack([a,b]))
```

### 차원추가
- np.newaxis 를 맨앞 또는 맨뒤에 넣어서 차원을 추가함.
- 텐서플로우 자체문법으로 tf.newaxis도 있음.
- 참고: 텐서와 파이토치자체에 넘파이와 비슷한 기능을 하는 명령어가 많음
    - 넘파이을 잘하면 텐서와 파이토치라는 인공지능 모듈에서 제공하는
      - 넘파이기능을 별도로 학습하지 않아도 됨
- np.expand와 같음

```python
a=np.arange(6)
print(a)
print(a[np.newaxis,:])  # a.reshape(1,6) 과 같음
print(a[:, np.newaxis]) # a.reshape(6,1)과 같음
```
