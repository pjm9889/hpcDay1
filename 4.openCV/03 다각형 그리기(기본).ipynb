```python
import cv2
import numpy as np

point_list=[] #    

def event_mous(event,x,y,flags,param):
    global point_list
    if event==cv2.EVENT_LBUTTONDOWN:# 
        point_list.append([x,y])
    if event==cv2.EVENT_LBUTTONDBLCLK:#   
        mask=np.zeros(img.shape, np.uint8)#  img
        point_list=np.array(point_list)#    
        mask=cv2.polylines(mask,[point_list],True,(255,255,255),3)#   
        mask2=cv2.fillPoly(mask,[point_list],(255,255,255))#    
        ROI=cv2.bitwise_and(img,mask2)#  , mask2   
        cv2.imwrite('ROI.png',ROI) #   ROI 
        cv2.imshow('roi',ROI) #Roi 
        

img=cv2.imread('./img02.jpg') #  
cv2.namedWindow('img')#     
cv2.setMouseCallback('img',event_mous,img)#  (setMouseCallback) envent_mous  
#, 
#:    ,  
cv2.imshow('img',img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
